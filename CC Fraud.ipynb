```r
library(caTools)
set.seed(23)
data_sample = sample.split(NewData$Class, SplitRatio = 0.8)
train_data = subset(NewData, data_sample==TRUE)
test_data = subset(NewData, data_sample==FALSE)
dim(train_data)
dim(test_data)
```

```r
log_model = glm(Class~., test_data, family=binomial())
```

```r
summary(log_model)
```

```r
plot(log_model)
```

```r
library(pROC)
lr.predict = predict(log_model, train_data, family=binomial())
auc.gbm = roc(test_data$Class, lr.predict, plot=TRUE, col="blue")
```

```r
install.packages("ROCR")
```

```r
library(pROC)
lr.predict = predict(log_model, train_data, probability=TRUE)
auc.gbm = roc(test_data$Class, lr.predict, plot=TRUE, col="blue")
```
